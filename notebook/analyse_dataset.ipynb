```python
import pandas as pd

# Import du dataset
df = pd.read_csv("https://raw.githubusercontent.com/WildCodeSchool/data-training-resources/main/quests/Market_retail.csv")
```

```python
df.info()
```

```python
# Vérification des doublons pour la colonne 'Order ID'
df['Order ID'].duplicated().sum()
```

```python
# Affichage des lignes avec des doublons dans la colonne 'Order ID'
doublons = df[df['Order ID'].duplicated(keep=False)]
doublons
```

On peut observer qu'une même commande peut contenir plusieurs lignes en fonction du nombre de produits différents commandés.

```python
# Identification des lignes strictement identiques
df[df.duplicated(keep=False)]
```

Il y'a deux lignes strictement identiques dans le dataset que nous ne garderons pas pour notre analyse.

```python
# Suppression des doublons
df = df.drop_duplicates()
```

Le reste des analyses sera fait sur Power BI.

```python
# Export pour analyse sur Power BI
df.to_csv("market_price_retail.csv", index=False)
```

Changement de format des colonnes (avec utilisation des paramètres régionaux):
- `Order Date` au format `date` + création de colonnes `mois` et `année` pour ajouter des filtres
- `Country` au format `pays`
- `State` au format `région`
- `City` au format `localité`
- `Discount` au format `nombre décimal` en supprimant le $ et en changeant les valeurs vides par 0
- `Sales` au format `nombre entier` en supprimant le $ + création d'une colonne contenant le montant après remise
- `Profit` au format `nombre entier` en supprimant le $ et en changeant les valeurs vides par 0
- `Quantity` au format `nombre entier`
- `Feedback ?` au format `bon`/`mauvais`

Vérification des données :
- `Region` contient uniquement 3 valeurs (`Central`, `North` et `South`)
- `Segment` contient uniquement 3 valeurs (`Consumer`, `Corporate` et `Home Office`)
- `Ship Mode` contient uniquement 4 valeurs (`Economy`, `Economy Plus`, `Immediate` et `Priority`)
- `Category` contient uniquement 3 valeurs (`Furniture`, `Office Supplies` et `Technology`)
